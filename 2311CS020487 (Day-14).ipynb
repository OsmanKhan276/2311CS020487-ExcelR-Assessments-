```python
df=sns.load_dataset('taxis')
df
```

```python
sns.barplot(x='color',y='total',data=df)
plt.show()   #This shows the graph of the attributes named total and color
```

```python
sns.barplot(x='color',y='total',data=df,color='yellow')
plt.show()   # color is used to change the color of the graph
```

```python
sns.barplot(x='color',y='total',data=df,estimator='mean')
plt.show()   #Estimator estimates the given value
```

```python
sns.histplot(x=df['fare'])  #hist is a histograph plot to know the distrubution of data
```

```python
sns.histplot(x=df['fare'],color='red')   #changes the color of the graph
```

```python
sns.histplot(x=df['fare'],color='grey',hue=df['color'])   #it includes the multiple colors for a graph
```

#### pallette is the method used to add the multiple colors to the graph 
#### for single color use color and for multiple comparable case use the palette
#### hue is the parameter for which the comparision is done

```python
sns.histplot(x=df['fare'],palette=['green','blue'],hue=df['payment'])  
```

```python
sns.violinplot(x=df['fare'],y=df['payment']) #this is the kinfd of graph for violin plotting
```

```python
sns.boxplot(x=df['fare']) #boxplot discovers the outliers in the data 
```

```python
sns.boxplot(x=df['fare'],y=df['tip'],palette=['red','green'])
```

```python
sns.scatterplot(x=df['fare'],y=df['total'],palette=['yellow','red'],hue=df['payment'])  #scatter plot is the category of relation plot
plt.show() #displays the graph
```

```python
sns.stripplot(x=df['color'],y=df['fare'],palette=['yellow','red'])
```
